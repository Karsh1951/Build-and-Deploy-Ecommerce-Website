```python
#importing the required libraries for data analysis
import numpy as np
import pandas as pd
#data visualization tools
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
#adding the style parameters to decorate the style of the plot(bg color)
sns.set_style('darkgrid')
plt.rcParams['font.size'] = 15 #runtime configuration parameters
plt.rcParams['figure.figsize']=(10,7)
#setting the face color 
plt.rcParams['figure.facecolor'] ='#FFE5B4'
```

```python
#LOADING THE DATA SET
data = pd.read_csv('/content/world-happiness-report-2021.csv')
#displaying the first five rows of the informaation
data.head(5)
```

```python
#dropping the columns that will not be used during the analysis
data_columns =['Country name','Regional indicator','Ladder score','Logged GDP per capita','Social support','Healthy life expectancy','Freedom to make life choices','Generosity','Perceptions of corruption']
```

```python
#copying the data columns to a new varata
print(list(data.columns))
```

```python
#copying the new data in a new var
data = data[data_columns].copy()


```

```python
#renaming columns to make it more simpler to to understand
happy_df = data.rename({'Country name': 'Country_name','Regional indicator': 'regional_indicator','Ladder score': 'happiness_score', 'Logged GDP per capita':'Logged_GDP_per_capita','Social support':'Social_support','Healthy life expectancy':'Healthy_life_expectancy','Freedom to make life choices':'Freedom_to_make_life_choices','Generosity':'Generosity','Perceptions of corruption':'Perceptions_of_corruption'}, axis=1)
```

```python
happy_df.head()
```

```python
#checking if any of the columns have a null value
happy_df.isnull().sum()
```

```python
from seaborn._core.properties import FontSize
#visualizing a plot between happiness score  and gdp regions
plt.rcParams['figure.figsize'] =(15,7) #SETTING RC PARAMETERS
plt.title('Plot between happiness_score and GDP') 
#CREATING A SCATTER PLOT,HUE IS PASSED IN FOR THE COLOR,s =giving the size of the dots
sns.scatterplot(x = happy_df.happiness_score, y=happy_df.Logged_GDP_per_capita,hue =happy_df.regional_indicator, s = 200);
#defining the legend-area describing the elements in graph(matplotlib)
plt.legend(loc='upper left',fontsize = '10')
plt.xlabel('Happiness score')
plt.ylabel('GDP Per Capita')

```

```python
#plot by pyplot to undestand gdp by region using group by function
gdp_region = happy_df.groupby('regional_indicator')['Logged_GDP_per_capita'].sum()
gdp_region
```

```python
#plotting  a pirchart for gdp by region(auto pct formats the value as a percentage with 1 decimal place)
gdp_region.plot.pie(autopct='%1.1f%%')
plt.title('GDP by region')
plt.ylabel
```

```python
#finding the total number of countries in each region
total_country = happy_df.groupby('regional_indicator')[['Country_name']].count()
print(total_country)
```

```python
#correlation map to depict the relationship between variables (peason correlation coefficient)
cor = happy_df.corr(method = "pearson")
f, ax= plt.subplots(figsize = (10, 5))
#drawing the heatmaps mask is passed tpo hide areas which has missing values atomaticalydata is not shown
sns.heatmap(cor, mask = np.zeros_like(cor,dtype=np.bool),
            cmap="Blues",square=True,ax=ax)
#color map = blue,displayed in square


```

```python
#visualizing a bar plot to tell the corruption rate in different regions mean fucntion is used to find the average corruption in each country
corruption = happy_df.groupby('regional_indicator')[['Perceptions_of_corruption']].mean()
corruption
```

```python
#visualizing the perceptions of corruption in a bargraph
plt.rcParams['figure.figsize']=(12,8)
plt.title('Perception of Corruption in varius regions')
#defining the x label
plt.xlabel('Region',fontsize = 15)
plt.ylabel('Corruption Index',fontsize = 15)
plt.xticks(rotation =30, ha='right')
plt.bar(corruption.index, corruption.Perceptions_of_corruption)
```

```python
#finding the life expectancy for top ten and bottom ten countries
top_10 = happy_df.head(10)
bottom_10 =happy_df.tail(10)
```

```python
#making the two visualizations 
fig, axes=plt.subplots(1,2, figsize =(16,6))
plt.tight_layout(pad= 2)
xlabels=top_10.Country_name
axes[0].set_title('Top 10 Happiest Countries life expectancy')
axes[0].set_xticklabels(xlabels, rotation=45,ha='right')
sns.barplot(x= top_10.Country_name,y=top_10.Healthy_life_expectancy,ax=axes[0])
axes[0].set_xlabel('Country Name')
axes[0].set_ylabel('Life expectancy')

#the bottom countries
xlabels=bottom_10.Country_name
axes[1].set_title('Bottom 10 least happy countries life Expectancies')
axes[1].set_xticklabels(xlabels, rotation=45,ha='right')
sns.barplot(x=bottom_10.Country_name,y= bottom_10.Healthy_life_expectancy,ax=axes[1])
axes[1].set_ylabel('life expectancy')
```

```python
#freedom to make life choices and the hapiness score using a scatter plot(hue parameter is used to differenciate diferent regions)
plt.rcParams['figure.figsize']=(15, 7)
sns.scatterplot(x=happy_df.Freedom_to_make_life_choices, y=happy_df.happiness_score,hue= happy_df.regional_indicator,s = 200)
plt.legend(loc = 'upper left',fontsize = '12')
plt.xlabel('Freedom to make life choices ')
plt.ylabel('Happines Score')

```

```python
#top ten most corrupt countries
country= happy_df.sort_values(by='Perceptions_of_corruption').head(10)
plt.rcParams['figure.figsize'] =(12, 6)
plt.xlabel('Country', fontsize=13)
plt.ylabel('Corruption Index',fontsize=13)
plt.xticks(rotation =30,ha='right')
plt.bar(country.Country_name,country.Perceptions_of_corruption)
```

```python
#scatter plot to visualize how corruption varies in terms of happiness score
plt.rcParams['figure.figsize']=(15, 7)
sns.scatterplot(x=happy_df.happiness_score, y=happy_df.Perceptions_of_corruption,hue=happy_df.regional_indicator,s=200)
plt.legend(loc='lower left',fontsize='14')
plt.xlabel('Happiness score')
plt.ylabel('Corruption')
```
